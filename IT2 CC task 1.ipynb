```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

converts it into a Pandas DataFrame, and adds the species as a column. It then displays:

1. The first few rows to preview the data.
2. Dataset information like column types and missing values.
3. Summary statistics, including mean, standard deviation, and range for the features.

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

print("First few rows:")
print(df.head())

print("\nDataset info:")
print(df.info())

print("\nSummary statistics:")
print(df.describe())
```

- **Histograms**: Display the frequency distribution of each feature (sepal and petal lengths and widths), helping to identify their spread and skewness.
- **Boxplots**: Show the spread of the data and highlight any potential outliers in the features.

```python

plt.figure(figsize=(10, 6))
df.hist(bins=20, figsize=(10, 8))
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df.iloc[:, :-1])
plt.title('boxplot of features')
plt.show()
```

At last **correlation heatmap** to visualize the relationships between numerical features. The heatmap displays the correlation coefficients, indicating how strongly features are linearly related:
- Values close to 1 indicate a strong positive correlation.
- Values close to -1 indicate a strong negative correlation.
- The `annot=True` option adds the exact correlation values to the heatmap for clarity.

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
